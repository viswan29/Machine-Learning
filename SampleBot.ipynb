```python
#preprocesssing:
def unique(list1):
    unique_list = [] 
    return [x for x in list1 if x not in unique_list]
```

```python
lemmatizer = WordNetLemmatizer()
def preprocessing(sent):
    rem_words = ['get', 'avail', 'who' , 'where', 'how' , 'what', 'why' , 'when', 'I', 'can']
    for p in list(punctuation):
        sent = sent.replace(p,'')
    sent = sent.lower().split()
    stop_words = set(stopwords.words('english'))
    sent = [w for w in sent if w not in stop_words]
    sent = [w for w in sent if w not in rem_words]
    sent = [lemmatizer.lemmatize(item) for item in sent]
    return(unique(sent))
```
